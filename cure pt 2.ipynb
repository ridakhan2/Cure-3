```python
import numpy as np
import pandas as pd 
from pandas import Series, DataFrame
import matplotlib.pyplot as plt 
```

```python
import sklearn 
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer(return_X_y = True, as_frame = True)
```

```python
a = cancer[0]
```

```python
b = cancer[1]
```

```python
a['classes'] = b
a
```

```python
a.shape
```

```python
a['types of cancer'] = b
```

```python
df = a.iloc[: , [0 ,2,3,31]]
df
```

```python
df = a.iloc[17:22,[0,2,3,31]]
df
```

```python
f0 = a[a['types of cancer'] == 0]
f1 = a[a['types of cancer'] == 1]
```

```python
fig, axs = plt.subplots(figsize = (12,3))

axs1 = plt.subplot2grid (shape = (1,4), loc = (0,0))
axs2 = plt.subplot2grid (shape = (1,4), loc = (0,1))
axs3 = plt.subplot2grid (shape = (1,4), loc = (0,2))
plt.tight_layout()

axs1.hist(f0["mean radius"], fc = 'none', edgecolor = 'b', label = 'c0(Malignant)')
axs1.hist(f1["mean radius"], fc = 'none', edgecolor = 'r', label = 'c1(Benign)')
axs1.set_xlabel('mean radius')
axs1.set_ylabel('Frequency')
axs1.legend()

axs2.scatter (f0["mean perimeter"], f0["mean radius"], label = 'c0(M)')
axs2.scatter (f1["mean perimeter"], f1["mean radius"], label = 'c1(B)')
axs2.set_xlabel('mean perimeter')
axs2.set_ylabel('mean radius')
axs2.legend()

axs3.scatter (f0["mean area"], f0["mean radius"], label = 'c0(M)')
axs3.scatter (f1["mean area"], f1["mean radius"], label = 'c1(B)')
axs3.set_xlabel('mean area')
axs3.set_ylabel('mean radius')
axs3.legend()
```
